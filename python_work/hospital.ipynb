```python
import csv


def showContent(content1):
    count = 0

    for line1 in content1:
        print(line1[0], line1[3], line1[6], sep = '/')
    
        count += 1
        if (count == 10):
            break
     
    
def findHosDep(content2):
    count = 0
    name = input("찾으려는 병원 과목을 입력하세요(ex 치과, 내과..) :")
    
    for line2 in content2:
        if name in line2[0]:
            print(line2)
            count += 1
            if(count == 10):
                break
    
    
def findHosInCity(content3):
    count = 0
    name = input("찾으려는 도시를 입력하세요(ex 서울, 인천, 대구..) : ")
    
    for line3 in content3:
        if name in line3[2]:
            print(line3)
            count += 1
            if(count == 10):
                break


f = open('hospital_2019.csv', 'r', encoding = 'utf8')
content = csv.reader(f)
header = next(content)

while True:
    print("1. 전체 병원/주소/의사수 목록")
    print("2. 병원 과목명 검색")
    print("3. 시도 병원 목록 출력")
    print("4. 종료")

    number = int(input("번호를 선택하세요 :"))
    if number == 1:
        showContent(content)
    elif number == 2:
        findHosDep(content)
    elif number == 3:
        findHosInCity(content)
    elif number == 4:
        break
    else:
        print("잘못 입력하셨습니다.")
```
